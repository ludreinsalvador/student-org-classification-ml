## Section 7: Classification Techniques - Decision Tree

```python
# Libraries for classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

print("\nLibraries for classification imported successfully!")
```

### Section 7.1: Classification with the Dataset

```python
# Dataset structure
print(f"\nOur dataset has {df.shape[0]} students with {df.shape[1]} features")
print("\nTarget variables we could predict:")
print("- is_active_member: Whether student is an active organization member")
print("- graduation_readiness: Student's readiness for graduation")
print("- leadership_experience: Whether student has leadership experience")

# Distribution of potential target variables
print("\n" + "-"*50)
print("DISTRIBUTION OF POTENTIAL TARGET VARIABLES:")
print("-"*50)

# Checking the is_active_member distribution
if 'is_active_member' in df.columns:
    print(f"\nActive Member Distribution:")
    print(df['is_active_member'].value_counts())
    print(f"Percentage breakdown:")
    print(df['is_active_member'].value_counts(normalize=True) * 100)

# Checking the graduation_readiness distribution  
if 'graduation_readiness' in df.columns:
    print(f"\nGraduation Readiness Distribution:")
    print(df['graduation_readiness'].value_counts())

# Checking the leadership_experience distribution
if 'leadership_experience' in df.columns:
    print(f"\nLeadership Experience Distribution:")
    print(df['leadership_experience'].value_counts())
```

### Section 7.2: Data Preparation

```python
# Using 'is_active_member' as the main classification target
# Binary classification problem in the student context

# Checking for any missing values in key columns
classification_features = [
    'age', 'year_level', 'gwa', 'units_enrolled', 'has_scholarship',
    'years_in_organization', 'attendance_rate', 'events_attended_last_year',
    'volunteer_hours_completed', 'extracurricular_engagement_score'
]

print("Checking for missing values in key features:")
for feature in classification_features:
    if feature in df.columns:
        missing_count = df[feature].isnull().sum()
        print(f"- {feature}: {missing_count} missing values")

# Creating a clean dataset for classification
df_clean = df.dropna(subset=['is_active_member'] + classification_features)
print(f"\nDataset after cleaning: {df_clean.shape[0]} rows")

# Preparing the features (X) and target (y)
X = df_clean[classification_features].copy()
y = df_clean['is_active_member'].copy()

print(f"\nFeatures shape: {X.shape}")
print(f"Target shape: {y.shape}")
print(f"\nTarget distribution after cleaning:")
print(y.value_counts())
print(f"\nTarget percentage distribution:")
print(y.value_counts(normalize=True) * 100)
```

### Section 7.3 Decision Tree Classifier - Part 1: Handle Categorical Data

```python
# Checking any categorical variables in the features
print("Checking data types in features:")
for col in X.columns:
    print(f"- {col}: {X[col].dtype}")

# Handle categorical variables if any exist (like year_level)
X_processed = X.copy()

# Label encode categorical variables
label_encoders = {}
categorical_columns = []

for column in X_processed.columns:
    if X_processed[column].dtype == 'object':
        categorical_columns.append(column)
        print(f"\nEncoding categorical column: {column}")
        print(f"Unique values: {X_processed[column].unique()}")
        
        le = LabelEncoder()
        X_processed[column] = le.fit_transform(X_processed[column])
        label_encoders[column] = le
        
        print(f"After encoding: {X_processed[column].unique()}")

if categorical_columns:
    print(f"\nCategorical columns processed: {categorical_columns}")
else:
    print("\nNo categorical columns found - all features are numerical")

print(f"\nProcessed features shape: {X_processed.shape}")
print("First 5 rows of processed features:")
print(X_processed.head())
```

### Section 7.3 Decision Tree Classifier - Part 2: Splitting the Data

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, 
    test_size=0.3, 
    random_state=42, 
    stratify=y
)

print(f"\nData split completed:")
print(f"Training set: {X_train.shape[0]} samples ({X_train.shape[0]/len(X_processed)*100:.1f}%)")
print(f"Testing set: {X_test.shape[0]} samples ({X_test.shape[0]/len(X_processed)*100:.1f}%)")

print(f"\nTraining set target distribution:")
print(y_train.value_counts())
print(f"\nTesting set target distribution:")
print(y_test.value_counts())
```

### Section 7.3 Decision Tree Classifier - Part 3: Training the Model

```python
print("\nTraining Decision Tree Classifier...")

# Initializing the Decision Tree with parameters to prevent overfitting
clf_tree = DecisionTreeClassifier(
    max_depth=5,          # Limit depth to prevent overfitting
    min_samples_split=10, # Minimum samples required to split
    min_samples_leaf=5,   # Minimum samples in leaf node
    random_state=42       # For reproducible results
)

# Training the model
clf_tree.fit(X_train, y_train)

print("✓ Decision Tree training completed!")
print(f"Tree depth: {clf_tree.get_depth()}")
print(f"Number of leaves: {clf_tree.get_n_leaves()}")
```

### Section 7.3 Decision Tree Classifier - Part 4: Predictions and Evaluation

```python
# Making predictions
y_pred_tree = clf_tree.predict(X_test)

# Calculating accuracy
accuracy_tree = accuracy_score(y_test, y_pred_tree)

print("DECISION TREE RESULTS")

print(f"Accuracy: {accuracy_tree:.4f} ({accuracy_tree*100:.2f}%)")

# Detailed classification report
print(f"\nDetailed Classification Report:")
print(classification_report(y_test, y_pred_tree))

# Confusion Matrix
print(f"\nConfusion Matrix:")
cm_tree = confusion_matrix(y_test, y_pred_tree)
print(cm_tree)
```

## Section 8: Classification Techniques - K-Nearest Neighbors (KNN) 

### Section 8.1 KNN - Part 1: Setup and Scaling

```python
print("KNN requires feature scaling since it's distance-based...")
print("Scaling features to have mean=0 and std=1")

# Initializing the scaler
scaler = StandardScaler()

# Scaling the training data and transform test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\nOriginal training data shape: {X_train.shape}")
print(f"Scaled training data shape: {X_train_scaled.shape}")

print(f"\nOriginal feature ranges (first 3 features):")
for i, col in enumerate(X_train.columns[:3]):
    print(f"- {col}: [{X_train.iloc[:, i].min():.2f}, {X_train.iloc[:, i].max():.2f}]")

print(f"\nScaled feature ranges (first 3 features):")
for i, col in enumerate(X_train.columns[:3]):
    print(f"- {col}: [{X_train_scaled[:, i].min():.2f}, {X_train_scaled[:, i].max():.2f}]")
```

### Section 8.2 KNN - Part 2: Finding Optimal K Value

```python
print(f"\nFinding optimal K value...")
print("Testing K values from 1 to 20...")

# Testing different k values
k_range = range(1, 21)
k_scores = []

# Calculating accuracy for each k value
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred_k = knn.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred_k)
    k_scores.append(accuracy)
    
    if k <= 5 or k % 5 == 0:  # Print progress for some k values
        print(f"K={k}: Accuracy = {accuracy:.4f}")

print("✓ K-value testing completed!")
```

### Section 8.3 KNN - Part 3: Visualize K Selection

```python
plt.figure(figsize=(12, 6))
plt.plot(k_range, k_scores, marker='o', linewidth=2, markersize=8)
plt.xlabel('K Value', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)
plt.title('KNN: Finding Optimal K Value', fontsize=14, fontweight='bold')
plt.xticks(k_range)
plt.grid(True, alpha=0.3)

# Highlighting the best k
best_k = k_range[np.argmax(k_scores)]
best_accuracy = max(k_scores)
plt.scatter(best_k, best_accuracy, color='red', s=100, zorder=5)
plt.annotate(f'Best K={best_k}\nAccuracy={best_accuracy:.4f}', 
             xy=(best_k, best_accuracy), xytext=(best_k+2, best_accuracy-0.01),
             arrowprops=dict(arrowstyle='->', color='red'), fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n" + "="*50)
print("OPTIMAL K VALUE RESULTS")
print("="*50)
print(f"Optimal K value: {best_k}")
print(f"Best accuracy with K={best_k}: {best_accuracy:.4f} ({best_accuracy*100:.2f}%)")
```

### Section 8.3 KNN - Part 4: Train Final Model

```python
print(f"\nTraining final KNN model with K={best_k}...")

# Training KNN with optimal k
knn_classifier = KNeighborsClassifier(n_neighbors=best_k)
knn_classifier.fit(X_train_scaled, y_train)

# Making predictions
y_pred_knn = knn_classifier.predict(X_test_scaled)

# Calculating accuracy
knn_accuracy = accuracy_score(y_test, y_pred_knn)

print("✓ KNN training completed!")
print("KNN MODEL RESULTS")
print(f"KNN Accuracy: {knn_accuracy:.4f} ({knn_accuracy*100:.2f}%)")

# Detailed classification report
print(f"\nDetailed Classification Report:")
print(classification_report(y_test, y_pred_knn, 
                          target_names=['Not Active Member', 'Active Member']))
```

### Section 8.3 KNN - Part 5: Confusion Matrix Visualization

```python
plt.figure(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_knn, cmap='Reds')
plt.title(f'KNN (K={best_k}) - Confusion Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Confusion matrix values
cm_knn = confusion_matrix(y_test, y_pred_knn)
print(f"\nConfusion Matrix Values:")
print(f"True Negatives (Not Active, Predicted Not Active): {cm_knn[0][0]}")
print(f"False Positives (Not Active, Predicted Active): {cm_knn[0][1]}")
print(f"False Negatives (Active, Predicted Not Active): {cm_knn[1][0]}")
print(f"True Positives (Active, Predicted Active): {cm_knn[1][1]}")
```
